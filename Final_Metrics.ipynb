```python
import matplotlib.pyplot as plt
import numpy as np

# Tipos de activación
activations = ['ReLU', 'LeakyReLU', 'Sigmoid', 'Tanh']

# Métricas
accuracy = [0.999420666409185, 0.9994382219725431, 0.9983673326077034, 0.999420666409185]
precision = [0.8658536585365854, 0.8505747126436781, 0.0, 0.8658536585365854]

x = np.arange(len(activations))
width = 0.35

plt.figure(figsize=(8,5))

# Barras
plt.bar(x - width/2, accuracy, width, label='Accuracy')
plt.bar(x + width/2, precision, width, label='Precision')

# Etiquetas
plt.xlabel('Tipo de activación')
plt.ylabel('Valor de la métrica')
plt.title('Comparación de funciones de activación')
plt.xticks(x, activations)

# Legend al lado
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

### Análisis Técnico

**1. ¿Qué función de activación obtuvo mejor precisión?**  
Las funciones que obtuvieron mejor precisión fueron la ReLU y Tanh, ya que ambas modelan relaciones de manera eficiente y ademas mostraron los mejores resultados en las métricas evaluadas.

**2. ¿Qué ocurre matemáticamente cuando no se utiliza activación?**  
Si no se usa una función de activación, la red neuronal solo realiza operaciones lineales simples. Esto hace que, aunque tenga muchas capas, el modelo funcione prácticamente como una sola capa y no pueda aprender patrones complejos de los datos.

**3. ¿Se observan indicios de saturación con Sigmoid o Tanh?**  
Sí. En Sigmoid se observa saturación cuando las salidas se acercan a 0 o 1, lo que produce gradientes muy pequeños (vanishing gradient) y dificulta el aprendizaje. En Tanh también puede ocurrir, pero en menor medida.

**4. ¿Existe diferencia práctica entre ReLU y Leaky ReLU?**  
Sí. ReLU puede generar neuronas muertas al anular valores negativos, mientras que Leaky ReLU mantiene un pequeño gradiente en esa región. Sin embargo, en este caso la diferencia en rendimiento fue mínima.

**5. ¿Cuál activación recomendarían para este problema y por qué?**  
Se recomienda la función de activación ReLU, ya que obtuvo buenas métricas de rendimiento y mostró predicciones más precisas en la matriz de confusión en comparación con las otras funciones.
